```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = 'https://www.amazon.in/s?k=toys&crid=1IAD3CN618MCE&sprefix=toys%2Caps%2C351&ref=nb_sb_ss_ts-doa-p_2_4'
```

```python
#Headers for request
HEADERS = ({'User_Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36', 'Accept_Language':'en-US, en,q= 0.5' })
```

```python
# HTTP request

webpage = requests.get(URL, headers=HEADERS)
```

```python
webpage
```

```python
webpage.content
```

```python
type(webpage.content)
```

```python
# Soup Object containiang all data
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
# Fetch links as List of Tag Objects
links = soup.find_all("a", attrs={'class':'a-size-base a-color-base a-link-normal a-text-normal'})
links
```

```python
links[0]
```

```python
links[0].get('href')
```

```python
link = links[0].get('href')
```

```python
product_list = "https://amazon.com" + link
```

```python
product_list
```

```python
new_webpage = requests.get(product_list, headers=HEADERS)
```

```python
new_webpage
```

```python
# Soup Object containiang all data
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
new_soup
```
